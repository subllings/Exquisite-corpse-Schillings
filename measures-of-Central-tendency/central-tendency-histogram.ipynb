## Mean Histograms


### The Mean - Dataset [1, 2, 2, 3, 4, 6, 9]

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
data = [1, 2, 2, 3, 4, 6, 9]
mean_val = np.mean(data)

# Plot
plt.figure(figsize=(8, 5))
plt.hist(data, bins=np.arange(0.5, 10.5, 1), edgecolor='black', color='skyblue', rwidth=0.9)
plt.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean ≈ {mean_val:.2f}')
plt.text(mean_val + 0.2, 1.8, f'Mean ≈ {mean_val:.2f}', color='red')
plt.title(r'$\mathbf{Data} = [1,\ 2,\ 2,\ 3,\ 4,\ 6,\ 9]$' + '\nHistogram of Data with Mean Indicator', fontsize=16)
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.yticks(fontsize=14)  
plt.xticks(np.arange(1, 10, 1), fontsize=14)

plt.legend()
#plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()



# Save and show
output_path = "/mnt/data/histogram_with_mean_and_data_label.png"
plt.savefig(output_path)
plt.show()

output_path
```

### How Outliers Can Distort the Mean

```python
import matplotlib.pyplot as plt
import numpy as np

# Example 1: Normal salaries
normal_salaries = [50_000, 52_000, 48_000, 51_000, 49_000, 50_000, 51_000]
mean_normal = np.mean(normal_salaries)

# Example 2: One outlier (CEO salary)
salaries_with_ceo = normal_salaries[:-1] + [1_000_000]  # replace one salary with CEO
mean_with_ceo = np.mean(salaries_with_ceo)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Graph 1: Normal salaries
axs[0].bar(range(len(normal_salaries)), normal_salaries, color='skyblue')
axs[0].axhline(mean_normal, color='red', linestyle='--', label=f'Mean = €{mean_normal:,.0f}')
axs[0].set_title('Without CEO Outlier')
axs[0].set_xlabel('Employee')
axs[0].set_ylabel('Salary (€)')
axs[0].legend()

# Graph 2: With CEO salary
axs[1].bar(range(len(salaries_with_ceo)), salaries_with_ceo, color='lightcoral')
axs[1].axhline(mean_with_ceo, color='red', linestyle='--', label=f'Mean = €{mean_with_ceo:,.0f}')
axs[1].set_title('With CEO Outlier')
axs[1].set_xlabel('Employee')
axs[1].set_ylabel('Salary (€)')
axs[1].legend()

# Save the image
plt.tight_layout()
output_path = "/mnt/data/mean_with_vs_without_ceo.png"
plt.savefig(output_path)
plt.show()

output_path
```

## Median Histograms 

### The Median - Dataset = [1,2,2,3,4,6,9]

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
data = [1, 2, 2, 3, 4, 6, 9]
median_value = np.median(data)

# Plotting
plt.figure(figsize=(8, 5))
plt.hist(data, bins=np.arange(min(data)-0.5, max(data)+1.5, 1), color='lightgreen', edgecolor='black')
plt.axvline(median_value, color='green', linestyle='--', linewidth=2, label=f'Median = {median_value:.1f}')
plt.title(r'$\mathbf{Data} = [1,\ 2,\ 2,\ 3,\ 4,\ 6,\ 9]$' + '\nHistogram of Data with Median Indicator', fontsize=16)


plt.xlabel('Value')
plt.ylabel('Frequency')

plt.yticks(fontsize=14)  
plt.xticks(np.arange(1, 10, 1), fontsize=14)

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Save the figure
output_path = "/mnt/data/histogram_median_with_dataset_corrected.png"
plt.savefig(output_path)
plt.show()

output_path


```

#### How the Median resists outliers

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset without outlier
data_without_ceo = [48000, 49000, 50000, 51000, 52000]
median_without = np.median(data_without_ceo)

# Dataset with outlier
data_with_ceo = [48000, 49000, 50000, 51000, 1_000_000]
median_with = np.median(data_with_ceo)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot without CEO outlier
axs[0].bar(range(len(data_without_ceo)), data_without_ceo, color='lightgreen')
axs[0].axhline(median_without, color='green', linestyle='--', label=f'Median = €{median_without:,.0f}')
axs[0].set_title('Without CEO Outlier')
axs[0].set_xlabel('Employee')
axs[0].set_ylabel('Salary (€)')
axs[0].legend()

# Plot with CEO outlier
axs[1].bar(range(len(data_with_ceo)), data_with_ceo, color='orange')
axs[1].axhline(median_with, color='green', linestyle='--', label=f'Median = €{median_with:,.0f}')
axs[1].set_title('With CEO Outlier')
axs[1].set_xlabel('Employee')
axs[1].set_ylabel('Salary (€)')
axs[1].legend()

# Save the image
plt.tight_layout()
output_path = "/mnt/data/median_with_vs_without_ceo.png"
plt.savefig(output_path)
plt.show()

output_path
```

### The Mode - Dataset = [1,2,2,3,4,6,9]

```python
import matplotlib.pyplot as plt
import numpy as np
from statistics import mode

# Dataset
data = [1, 2, 2, 3, 4, 6, 9]

# Histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=np.arange(min(data), max(data) + 2) - 0.5, color='lightcoral', edgecolor='black')

# Mode calculation and vertical line
mode_value = mode(data)
plt.axvline(mode_value, color='blue', linestyle='--', linewidth=2, label=f'Mode = {mode_value}')

# Set x-axis ticks to show all integers from 1 to 9
plt.xticks(np.arange(1, 10, 1))

# Titles and labels
plt.title(r'$\mathbf{Data} = [1,\ 2,\ 2,\ 3,\ 4,\ 6,\ 9]$' + '\nHistogram of Data with Mode Indicator', fontsize=16)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.yticks(fontsize=14)  
plt.xticks(np.arange(1, 10, 1), fontsize=14)

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Save and return the path
output_path = "/mnt/data/mode_histogram_corrected_full_ticks.png"
plt.savefig(output_path, dpi=300)
plt.show()

output_path



```

### The Mode - WHEN TO USE THE MODE Effectively

```python
import matplotlib.pyplot as plt
from collections import Counter

# --- 1. Pizza Toppings (Categorical Data) ---
pizza_data = ['Cheese', 'Cheese', 'Pepperoni', 'Cheese', 'Veggie', 'Pepperoni', 'Cheese']
pizza_counts = Counter(pizza_data)

plt.figure(figsize=(6, 4))
plt.bar(pizza_counts.keys(), pizza_counts.values(), color='lightcoral')
plt.title("Most Preferred Pizza Toppings")
plt.xlabel("Topping")
plt.ylabel("Frequency")
plt.axhline(max(pizza_counts.values()), linestyle='--', color='red', label='Mode = Cheese')
plt.legend()
plt.tight_layout()
pizza_path = "/mnt/data/pizza_mode_chart.png"
plt.savefig(pizza_path)
plt.show()

# --- 2. Shoe Sizes (Numerical Data) ---
shoe_sizes = [36, 37, 38, 38, 38, 39, 40]
shoe_counts = Counter(shoe_sizes)

plt.figure(figsize=(6, 4))
plt.bar(shoe_counts.keys(), shoe_counts.values(), color='lightblue')
plt.title("Most Sold Shoe Size")
plt.xlabel("Shoe Size")
plt.ylabel("Sales Count")
plt.axvline(38, color='blue', linestyle='--', label='Mode = 38')
plt.legend()
plt.tight_layout()
shoes_path = "/mnt/data/shoes_mode_chart.png"
plt.savefig(shoes_path)
plt.show()

pizza_path, shoes_path
```

### Summary Table: How Central Measures Differ

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import MaxNLocator

# Define datasets
datasets = {
    "Outlier (1,2,2,3,100)": [1, 2, 2, 3, 100],
    "Symmetric (10,10,10,10,10)": [10, 10, 10, 10, 10],
    "Skewed (1,1,2,3,4,5)": [1, 1, 2, 3, 4, 5]
}

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle("Visualizing Different Data Distributions", fontsize=14)

for ax, (title, data) in zip(axs, datasets.items()):
    sns.histplot(data, bins=np.arange(min(data), max(data)+2)-0.5, kde=False, ax=ax, color='skyblue')
    ax.set_title(title)
    ax.set_xlabel("Value")
    ax.set_ylabel("Frequency")
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))
    ax.set_ylim(0, max(np.bincount(data)) + 1)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
output_path = "/mnt/data/central_tendency_distributions.png"
plt.savefig(output_path)
plt.show()

output_path
```
